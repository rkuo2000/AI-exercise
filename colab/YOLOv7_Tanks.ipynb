## Prepare Datasets


```python
import kagglehub
rkuo2000_tank_by_s_danish_path = kagglehub.dataset_download('rkuo2000/tank-by-s-danish')
print(rkuo2000_tank_by_s_danish_path)

import os
os.listdir(rkuo2000_tank_by_s_danish_path)
```

## Create data.yaml

```python
%%writefile data/tanks.yaml

train: /kaggle/input/tank-by-s-danish/train/images
val: /kaggle/input/tank-by-s-danish/valid/images
test: /kaggle/input/tank-by-s-danish/test/images

nc: 3
names: ['tank', 'vehicle', 'wqw']
```

## Download YOLOv7 pretrained model

```python
!wget https://github.com/WongKinYiu/yolov7/releases/download/v0.1/yolov7-tiny.pt
```

```python
!sed -i "s/nc: 80/nc: 3/" cfg/training/yolov7-tiny.yaml
!sed -i "s/IDetect/Detect/" cfg/training/yolov7-tiny.yaml
```

## Train Model

```python
!python train.py --epochs 300 --workers 4 --device 0 --batch-size 32 \
--data data/tanks.yaml --img 640 640 --cfg cfg/training/yolov7-tiny.yaml \
--weights 'yolov7-tiny.pt' --name exp --hyp data/hyp.scratch.tiny.yaml
```
